```python
def valor_formulas(string):
  auxiliar = []
  string = string.replace(" ", "")

  i = 0
  teste = 0

  while i < len(string):
      if string[i] == '=':
        teste = 1

      if teste == 1 and (string[i] == ',' or string[i] == '='):
        auxiliar = string[:i + 1] + 'F' + string[i + 1:]
        string = auxiliar

      if teste == 1 and i == 0:
        auxiliar = 'F' + string[i + 1:]
        auxiliar = auxiliar[1:]
        string = auxiliar

      if teste == 0 and i == 0:
        auxiliar = 'T' + string
        string = auxiliar

      if teste == 0 and string[i] == ',':
        auxiliar = string[:i + 1] + 'T' + string[i + 1:]
        string = auxiliar

      i = i + 1

  return string


def conectivo_principal(string, inicio, fim):
  pilha = []

  string = string.replace(" ", "")

  i = inicio
  conectivo = None
  posicao = 0
  auxiliar = []

  while(i < len(string)):
    if(not(string[i] >= 'a' and string[i] < 'z')):
      pilha.append(string[i])

    if string[i] == '~' or string[i] == 'V' or string[i] == '&' or string[i] == '>':
      auxiliar.append(i)
    if pilha[-1] == ')':
      if (pilha[-2] == '~' or pilha[-2] == 'V' or pilha[-2] == '&' or pilha[-2] == '>') and pilha[-3] == '(':
        conectivo = pilha[-2]
        posicao = auxiliar[-1]
        auxiliar.pop()
        pilha.pop()
        pilha.pop()
        pilha.pop()
    i = i + 1
  return [conectivo, posicao]



def qtd_betas(ramo):

  beta = ''

  if(verifica_atomo(ramo)):
    beta = 'atomo'
  else:
    conec_aux = conectivo_principal(ramo, 1, len(ramo))
    if ramo[0] == 'T':
      if conec_aux[0] == 'V' or conec_aux[0] == '>':
        beta = 'BETA'
      elif conec_aux[0] == '&' or conec_aux[0] == '~':
        beta = 'ALFA'
    elif ramo[0] == 'F':
      if conec_aux[0] == '&':
        beta = 'BETA'
      elif conec_aux[0] == 'V' or conec_aux[0] == '>' or conec_aux[0] == '~':
        beta = 'ALFA'

  return beta



def cria_ramo(string):
  anterior = 0
  j = 0
  tam_atual = 0
  ramo = []

  for i in range(len(string)):
    if string[i] == ',' or string[i] == '=' or i == len(string) - 1:
      ramo.append(string[anterior:i])
      anterior = i + 1
      j = j + 1
      tam_atual = tam_atual + 1

  ramo[-1] = ramo[-1] + ')'
  return ramo

def verifica_atomo(string):
  return len(string) == 4

def ramo_fechado(ramo):
  if not(verifica_atomo(ramo[-1])):
    return False
  else:
    for i in range(len(ramo[:-1])):
      if verifica_atomo(ramo[i]):
        if ramo[-1][0] != ramo[i][0] and ramo[-1][1:] == ramo[i][1:]:
          return True

  return False

print('Símbolos: (¬ := ~), (∧ := &), (∨ := V), (→ := >), (⊢ := =)')
print('Lembre-se de colocar todos os parênteses necessários.')
print()
string = input("Digite o conjunto de fórmulas: ")
string = valor_formulas(string)

ramo = cria_ramo(string)
beta = []
PilhaDeRamos = []

for i in range(len(ramo)):
  beta.append(qtd_betas(ramo[i]))

TamAtual = len(ramo)

#Expanções alfas
while True:
  i = 0
  formula_valida = True
  TamAtual = len(ramo)

  while i < TamAtual and beta[i] != 'ALFA':
    i = i + 1


  if i < TamAtual:
    auxiliar = conectivo_principal(ramo[i], 1, len(ramo[i]) - 1)
    if ramo[i][0] == 'T':
      if auxiliar[0] == '&':
        if len(ramo[i][1:auxiliar[1]]) == 2:
          ramo.append('T' + ramo[i][1:auxiliar[1]] + ')')
        else:
          ramo.append('T' + ramo[i][2:auxiliar[1]])

        if len(ramo[i][auxiliar[1] + 1:len(ramo[i])]) == 2:
          ramo.append('T' + '(' + ramo[i][auxiliar[1] + 1:len(ramo[i])])
        else:
          ramo.append('T' + ramo[i][auxiliar[1] + 1:len(ramo[i]) - 1])

        beta[i] = 'alfa'

      elif auxiliar[0] == '~':
        if len(ramo[i][auxiliar[1] + 1:len(ramo[i])]) == 2:
          ramo.append('F' + '(' + ramo[i][auxiliar[1] + 1:len(ramo[i])])
        else:
          ramo.append('F' + ramo[i][auxiliar[1] + 1:len(ramo[i]) - 1])
        beta[i] = 'alfa'

    elif ramo[i][0] == 'F':
      if auxiliar[0] == 'V':
        if len(ramo[i][1:auxiliar[1]]) == 2:
          ramo.append('F' + ramo[i][1:auxiliar[1]] + ')')
        else:
          ramo.append('F' + ramo[i][2:auxiliar[1]])

        if len(ramo[i][auxiliar[1] + 1:len(ramo[i])]) == 2:
          ramo.append('F' + '(' + ramo[i][auxiliar[1] + 1:len(ramo[i])])
        else:
          ramo.append('F' + ramo[i][auxiliar[1] + 1:len(ramo[i]) - 1])
        beta[i] = 'alfa'

      elif auxiliar[0] == '>':
        if len(ramo[i][1:auxiliar[1]]) == 2:
          ramo.append('T' + ramo[i][1:auxiliar[1]] + ')')
        else:
          ramo.append('T' + ramo[i][2:auxiliar[1]])
        if len(ramo[i][auxiliar[1] + 1:len(ramo[i])]) == 2:
          ramo.append('F' + '(' + ramo[i][auxiliar[1] + 1:len(ramo[i])])
        else:
          ramo.append('F' + ramo[i][auxiliar[1] + 1:len(ramo[i]) - 1])
        beta[i] = 'alfa'

      elif auxiliar[0] == '~':
        if len(ramo[i][auxiliar[1] + 1:len(ramo[i])]) == 2:
          ramo.append('T' + '(' + ramo[i][auxiliar[1] + 1:len(ramo[i])])
        else:
          ramo.append('T' + ramo[i][auxiliar[1] + 1:len(ramo[i]) - 1])
        beta[i] = 'alfa'

    if auxiliar[0] != '~':
      beta.append(qtd_betas(ramo[-2]))
    beta.append(qtd_betas(ramo[-1]))

    TamAtual = len(ramo)

  else:

    i = 0

    while i < TamAtual and beta[i] != 'BETA':

      i = i + 1

    if i < TamAtual:
      auxiliar = conectivo_principal(ramo[i], 1, len(ramo[i]) - 1)
      if ramo[i][0] == 'F':
        if auxiliar[0] == '&':
          if len(ramo[i][1:auxiliar[1]]) == 2:
            ramo.append('F' + ramo[i][1:auxiliar[1]] + ')')
          else:
            ramo.append('F' + ramo[i][2:auxiliar[1]])

          beta.append(qtd_betas(ramo[-1]))
          TamAtual = len(ramo)

          if len(ramo[i][auxiliar[1] + 1:len(ramo[i])]) == 2:
            PilhaDeRamos.append(['F' + '(' + ramo[i][auxiliar[1] + 1:len(ramo[i])], len(ramo), beta])
          else:
            PilhaDeRamos.append(['F' + ramo[i][auxiliar[1] + 1:len(ramo[i]) - 1], len(ramo), beta])
          beta[i] = 'beta'

      elif ramo[i][0] == 'T':
        if auxiliar[0] == 'V':
          if len(ramo[i][1:auxiliar[1]]) == 2:
            ramo.append('T' + ramo[i][1:auxiliar[1]] + ')')
          else:
            ramo.append('T' + ramo[i][2:auxiliar[1]])

          beta.append(qtd_betas(ramo[-1]))
          TamAtual = len(ramo)

          if len(ramo[i][auxiliar[1] + 1:len(ramo[i])]) == 2:
            PilhaDeRamos.append(['T' + '(' + ramo[i][auxiliar[1] + 1:len(ramo[i])], len(ramo), beta])
          else:
            PilhaDeRamos.append(['T' + ramo[i][auxiliar[1] + 1:len(ramo[i]) - 1], len(ramo), beta])
          beta[i] = 'beta'

        elif auxiliar[0] == '>':
          if len(ramo[i][1:auxiliar[1]]) == 2:
            ramo.append('F' + ramo[i][1:auxiliar[1]] + ')')
          else:
            ramo.append('F' + ramo[i][2:auxiliar[1]])

          beta.append(qtd_betas(ramo[-1]))
          TamAtual = len(ramo)

          if len(ramo[i][auxiliar[1] + 1:len(ramo[i])]) == 2:
            PilhaDeRamos.append(['T' + '(' + ramo[i][auxiliar[1] + 1:len(ramo[i])], len(ramo), beta])
          else:
            PilhaDeRamos.append(['T' + ramo[i][auxiliar[1] + 1:len(ramo[i]) - 1], len(ramo), beta])
          beta[i] = 'beta'

      TamAtual = len(ramo)

    else:
      if len(ramo) == 1:
        verifica_ramo = ramo_fechado(ramo)
      else:
        verifica_ramo = ramo_fechado(ramo[:-1])
        if not verifica_ramo:
          verifica_ramo = ramo_fechado(ramo)

      if not verifica_ramo:
        formula_valida = False
        break

      if len(PilhaDeRamos) != 0:
        PilhaAuxiliar = PilhaDeRamos.pop()
        beta  = PilhaAuxiliar[2]
        ramo = ramo[:PilhaAuxiliar[1] - 1]
        ramo.append(PilhaAuxiliar[0])

      else:
        break
  TamAtual = len(ramo)
  if len(ramo) > 1:
    verifica_ramo = ramo_fechado(ramo[:-1])
    if not verifica_ramo:
      verifica_ramo = ramo_fechado(ramo)

  if verifica_ramo:
    if len(PilhaDeRamos) != 0:
      PilhaAuxiliar = PilhaDeRamos.pop()
      beta  = PilhaAuxiliar[2]
      ramo = ramo[:PilhaAuxiliar[1] - 1]
      ramo.append(PilhaAuxiliar[0])
  else:
    i = 0
    while i < TamAtual and beta[i] != 'ALFA':
      i = i + 1
    if i > TamAtual:
      break
    i = 0
    while i < TamAtual and beta[i] != 'BETA':
      i = i + 1
    if i > TamAtual:
      break

if formula_valida:
  print("Fórmula válida.")
else:
  print("Não é possível garantir a validade.")
```
